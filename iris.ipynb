```python
# 2. Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# 3. Import Data
iris= pd.read_csv("/Users/asif/Desktop/Python Iris and mtcars/Iris Python/Iris.csv")
```

```python
#4. Varialbe names
iris.head()
iris.tail()

```

```python
# Two options to specify variable name: 
# First, file name, followed by variable name: iris['Sepal.Length']
# Second, define speal length as a variable; e.g., sepal_length = iris['Sepal.Length]. In this case, you
# only use sepal_length 
```

```python
sepal_length = iris['Sepal.Length']
```

```python
# Example:
mean = np.mean(iris['Sepal.Length'])
mean = np.mean(sepal_length)
```

```python
# Descriptive Statistics
mean = np.mean(sepal_length)
print(mean)
print("Mean:", mean)
median=np.median(sepal_length)
print("Median:", median)
std_dev=np.std(sepal_length)
print("Standard Deviation:", std_dev)
mode = sepal_length.mode()
print("Mode:", mode)
Q1 = np.percentile(sepal_length, 25)
print("First Quartile:", Q1)
Q2= np.percentile(sepal_length, 50)
print("Second Quartile:", Q2)
Q3 = np.percentile(sepal_length, 75)
print("Third Quartile:", Q3)
```

```python
# Descriptive Statistics
mean = np.mean(sepal_length)

print(mean)
```

```python
# Descriptive Statistics
descriptive = sepal_length.describe()
print(descriptive)
```

```python
# Histogram --- plt.hist(variable name)
plt.hist(sepal_length, bins=7)
plt.xlabel("Sepal Length")
plt.ylabel("Count")
plt.title("Histogram of Sepal Length")
```

```python
# Boxplot -- plt.boxplot (variable name)
plt.boxplot(sepal_length)
plt.title("Boxplot of Sepal Length")
```

```python
# Scatter Plot # Basic code = plt.scatter(x-variable, y-variable)
# First method
plt.scatter(iris['Sepal.Length'], iris['Petal.Length'])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
```

```python
# Scatter Plot # Basic code = plt.scatter(x-variable, y-variable)
# Second method
sepal_length = iris['Sepal.Length']
petal_length= iris['Petal.Length']
plt.scatter(sepal_length, petal_length)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Scatter Plot")
```

```python
# Violin Plot -- code: plt.violinplot(variable name)
plt.violinplot(sepal_length)
plt.xlabel("Density")
plt.ylabel("Sepal Length")
plt.title("Violin Plot of Sepal Length")
```

```python
# Line Graph -- code: plt.plot(varaible name, marker = 'o', linestyle = "-")
plt.plot(sepal_length, marker = "o", linestyle = "-")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.title("Line Graph of Sepal Length")
```

```python
# Bar plot - 2 Step process
mean_sepal_length = iris['Sepal.Length'].mean()
mean_sepal_width = iris['Sepal.Width'].mean()
mean_petal_length = iris['Petal.Length'].mean()
mean_petal_width = iris['Petal.Width'].mean()
```

```python
iris.head()
iris.tail()
```

```python
labels = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
means = [mean_sepal_length, mean_sepal_width, mean_petal_length, mean_petal_width]
plt.bar(labels, means) # This one also works perfect. but for aesthetics add colors.
plt.bar(labels, means, color = ['turquoise', 'green', 'red', 'pink'])
```

```python
# Pie Chart - 2 step process
species_counts = iris['Species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'], startangle=140)
```
